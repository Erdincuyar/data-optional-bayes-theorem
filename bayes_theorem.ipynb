## 2) Prior probability : $ \mathbb{P}(play)$

ğŸ¤” Bir oyunun oynanmasÄ±nÄ±n teorik olasÄ±lÄ±ÄŸÄ± nedir â“

Bunu, Ã¶nceki tablondaki sayÄ±lara bakarak bulabilirsin.

<details>
    <summary>Cevap</summary>
    
| Weather  | Played | No | Total |
|----------|------|----|-------|
| Total    | 9    | 5  | 14    |     
    
$ \mathbb{P}(played) = \frac{9}{14} = 64.29 \% $
</details>

ğŸ‘©ğŸ»â€ğŸ’» `prior_probability` fonksiyonunu yazarak sonucu hesapla.

````python
def prior_probability(event_name: str, observations: list) -> float:
    if not observations:
        return 0.0

    # observations listesi iÃ§indeki event_name ("Yes") sayÄ±sÄ±nÄ± bulur
    event_count = observations.count(event_name)
    total_count = len(observations)

    return event_count / total_count

# --- TEST VERÄ°SÄ° ---
# Tablodaki verilere gÃ¶re: 9 adet "Yes", 5 adet "No"
play_data_example = (["Yes"] * 9) + (["No"] * 5)

# ğŸ‘€ KODUNU test etmek iÃ§in aÅŸaÄŸÄ±dakini Ã§alÄ±ÅŸtÄ±r.
# ArtÄ±k "Yes" metni ile listedeki "Yes" metinlerini doÄŸru eÅŸleÅŸtiriyoruz.
assert(round(prior_probability("Yes", play_data_example), 4) == 0.6429)# ğŸ‘€ KODUNU test etmek iÃ§in aÅŸaÄŸÄ±dakini Ã§alÄ±ÅŸtÄ±r.
#EÄŸer hiÃ§bir ÅŸey gÃ¶rÃ¼nmÃ¼yorsa, fonksiyonun doÄŸru Ã§alÄ±ÅŸÄ±yor demektir.Aksi halde, kodunu dÃ¼zeltmek iÃ§in inceleme yap!
assert(round(prior_probability("Yes", play_data_example),4) == 0.6429)
````

## 3) Likelihood :  $ \mathbb{P}(weather | play)$

ğŸ¤” Bir oyunun OYNANMAMIÅ olduÄŸunu bildiÄŸimizde havanÄ±n yaÄŸmurlu olma olasÄ±lÄ±ÄŸÄ± nedir?

Bunu Ã¶nceki frekans tablosundan okuyabilirsin.

<details>
    <summary>Cevap</summary>
    
| Weather  | No | 
|----------|----|
| Sunny    | 2  | 
| Overcast | 0  | 
| Rainy    | 3  | 
| Total    | 5  |         
    
$ \mathbb{P}(play) = \frac{3}{5} = 60 \% $
</details>

````python
def likelihood(weather, played, weather_data, play_data):
    # 'played' durumuna sahip olanlarÄ±n iÃ§indeki 'weather' sayÄ±sÄ±nÄ± bul
    
    # AdÄ±m 1: Sadece 'played' (Ã¶rn: "No") olan indekslerdeki hava durumlarÄ±nÄ± filtrele
    filtered_weather = [w for w, p in zip(weather_data, play_data) if p == played]
    
    # AdÄ±m 2: FiltrelenmiÅŸ liste boÅŸsa 0 dÃ¶ndÃ¼r, deÄŸilse oranÄ± hesapla
    if not filtered_weather:
        return 0.0
    
    return filtered_weather.count(weather) / len(filtered_weather)

# --- TABLOYA GÃ–RE VERÄ°LERÄ° OLUÅTURALIM (14 GÃœN) ---
# Sunny: 3 Played, 2 No
# Overcast: 4 Played, 0 No
# Rainy: 2 Played, 3 No

weather_data_example = (
    ["Sunny"] * 5 + 
    ["Overcast"] * 4 + 
    ["Rainy"] * 5
)

play_data_example = (
    ["Yes", "Yes", "Yes", "No", "No"] +    # Sunny (3Y, 2N)
    ["Yes", "Yes", "Yes", "Yes"] +         # Overcast (4Y, 0N)
    ["Yes", "Yes", "No", "No", "No"]       # Rainy (2Y, 3N)
)

# ğŸ‘€ TEST: P(Rainy | No) = 3 / 5 = 0.60
assert(likelihood("Rainy", "No", weather_data_example, play_data_example) == 0.60)
````

## 4) Posterior probability : $ \large P(play \mid weather ) = P(play) \times \frac{P(weather \mid play)}{P(weather)} $

ğŸ”¥ Sonunda posterior olasÄ±lÄ±ÄŸÄ± hesaplayabiliriz:

$$\large \text{posterior probability} = \text{prior probability} \times \text{likelihood} \times \beta $$ 

where $ \large \beta = \frac{1}{P(weather)} $ is a _normalization factor_.
 
<details>
    <summary>Beklenen sonuÃ§lar</summary>

Ã–nceden doldurduÄŸun tabloyu hatÄ±rlÄ±yor musun?

| Weather  | Played | No | Total |
|----------|------|----|-------|
| Sunny    | 3    | 2  | 5     |
| Overcast | 4    | 0  | 4     |
| Rainy    | 2    | 3  | 5     |
| Total    | 9    | 5  | 14    |   
    
Bu tabloya gÃ¶re $ \mathbb{P}(played | weather) $ ÅŸÃ¶yle:
    
| Weather  | Proba(Played\|Weather) | Proba(No\|Weather) |
|----------|----------------------|--------------------|
| Sunny    | 3/5 = 0.6                  | 2/5 = 0.4                |
| Overcast | 4/4 = 1.0                  | 0/4 = 0.0                |
| Rainy    | 2/5 = 0.4                  | 3/5 = 0.6                |
    
</details>

````python
def posterior_probability(played, weather, weather_data, play_data):
    """
    P(played | weather) = ( P(played) * P(weather | played) ) / P(weather)
    """
    # 1. Prior Probability: P(played)
    # Toplamda oyunun oynanma (veya oynanmama) genel olasÄ±lÄ±ÄŸÄ±
    p_prior = prior_probability(played, play_data)
    
    # 2. Likelihood: P(weather | played)
    # Oyunun oynandÄ±ÄŸÄ± (veya oynanmadÄ±ÄŸÄ±) bilindiÄŸinde havanÄ±n o durumda olma olasÄ±lÄ±ÄŸÄ±
    p_likelihood = likelihood(weather, played, weather_data, play_data)
    
    # 3. Normalization Factor (Beta): 1 / P(weather)
    # HavanÄ±n o durumda olma genel olasÄ±lÄ±ÄŸÄ±
    p_weather = prior_probability(weather, weather_data)
    
    if p_weather == 0:
        return 0.0
    
    beta = 1 / p_weather
    
    # Posterior FormÃ¼lÃ¼: Prior * Likelihood * Beta
    posterior = p_prior * p_likelihood * beta
    
    return posterior

# ğŸ‘€ Test Edelim: Tabloya gÃ¶re P(Played="Yes" | Weather="Sunny") = 3/5 = 0.6 olmalÄ±
# Ya da P(Played="No" | Weather="Rainy") = 3/5 = 0.6 olmalÄ±

result = posterior_probability("Yes", "Sunny", weather_data_example, play_data_example)
print(f"P(Yes | Sunny) sonucu: {result}")

assert(round(posterior_probability("Yes", "Sunny", weather_data_example, play_data_example), 1) == 0.6)
assert(round(posterior_probability("No", "Rainy", weather_data_example, play_data_example), 1) == 0.6)
````

````python
# ğŸ‘€ AÅŸaÄŸÄ±dakini Ã§alÄ±ÅŸtÄ±rarak kodunu test et.
assert(np.isclose(posterior_probability("Yes", "Sunny", weather_data_example, play_data_example), 0.6))
assert(np.isclose(posterior_probability("No", "Sunny", weather_data_example, play_data_example), 0.4))
assert(np.isclose(posterior_probability("Yes", "Overcast", weather_data_example, play_data_example), 1.0))
assert(np.isclose(posterior_probability("No", "Overcast", weather_data_example, play_data_example), 0))
assert(np.isclose(posterior_probability("Yes", "Rainy", weather_data_example, play_data_example), 0.4))
assert(np.isclose(posterior_probability("No", "Rainy", weather_data_example, play_data_example), 0.6))
````
